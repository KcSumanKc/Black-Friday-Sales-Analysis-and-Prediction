### Data Preprocessing

```python
df
```

```python
df['Product_ID'] = df['Product_ID'].str.replace('P00', '')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
df['Product_ID'] = ss.fit_transform(df['Product_ID'].values.reshape(-1,1))
```

```python
df.drop(['Product_Category_3'], axis =1, inplace = True)
```

```python
df['Product_Category_2']= df['Product_Category_2'].fillna(df['Product_Category_2'].mean())
```

```python
df.isna().sum()
```

```python
df.head()
```

### Label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
categorical_columns = ['Gender', 'City_Category', 'Age']
le = LabelEncoder()

for i in categorical_columns:
    df[i] = le.fit_transform(df[i])
df.dtypes
    
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].replace('4+', '4')
```

```python
df['Gender'] = df['Gender'].astype(int)
df['Age'] = df['Age'].astype(int)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
df['City_Category'] = df['City_Category'].astype('category')
```

```python
df.head()
```

### Distribution Plots

```python
rows = 3
cols =3
fig , ax = plt.subplots(nrows = rows, ncols = cols, figsize =(10,4))
col = df.columns
index =2
#print(cols)
for i in range(rows):
    for j in range(cols):
        sns.displot(df[col[index]], ax=ax[i][j])
        index = index+1
plt.tight_layout()   
```
